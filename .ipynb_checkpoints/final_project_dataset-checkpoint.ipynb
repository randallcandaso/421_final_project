```python
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

amd = pd.read_csv('amd.csv')
intel = pd.read_csv('intel.csv')
```

```python
# combining the different cache types into one, using the highest number type as preference

amd['cache_size'] = amd['cache_l3'].fillna(amd['cache_l2']).fillna(amd['cache_l1']).fillna(0)
amd = amd.drop(columns=['cache_l3', 'cache_l2', 'cache_l1'])
```

<span style="color:blue"><b>Let's check the correlation of our features.</b></span>

```python
numerical_amd = amd.select_dtypes(include=['int64', 'float64'])
numerical_amd.corr()
```

```python
def group_product_line_amd(product_line):
    '''

    product_line: name/model of processor (str)
    return: tier of processor (str)

    AMD CPU's are grouped into performance tiers based off their product lines.

    '''
    line = product_line.lower()
    words = line.split()
    
    if "epyc" in line:
        return "High Tier"
    elif "threadripper" in line:
        return "High Tier"
    elif "ryzen" in line and "pro" in words:
        return "High Tier"
    elif "ryzen" in line and "9" in words:
        return "High Tier"
    
    elif "ryzen" in line and "mobile" in line:
        return "Mid Tier"
    elif "ryzen" in line and ("5" in words or "7" in words):
        return "Mid Tier"
    
    elif "fx" in line:
        return "Low Tier"
    elif "athlon" in line:
        return "Low Tier"
    elif "phenom" in line:
        return "Low Tier"
    elif "opteron" in line:
        return "Low Tier"
    elif "a4" in line or "a6" in line or "a8" in line or "a9" in line or "a10" in line or "a12" in line:
        return "Low Tier"
    elif "e1" in line or "e2" in line:
        return "Low Tier"
    elif "business class" in line:
        return "Low Tier"
    
    else:
        return "Other"
```

```python
def group_product_line_intel(product_line):
    '''

    product_line: name/model of processor (str)
    return: tier of processor (str)

    Intel CPU's are grouped into performance tiers based off their product lines.

    '''
    line = product_line.lower()
    words = line.split()

    if "xeon" in line or "x-series" in line:
        return "High Tier"
    
    elif "atom" in line or "celeron" in line or "pentium gold" in line or "pentium silver" in line:
        return "Low Tier"
        
    elif "2nd generation" in line or "3rd generation" in line or "4th generation" in line or "5th generation" in line or "6th generation" in line or "7th generation" in line or "8th generation" in line or "9th generation" in line or "10th generation" in line or "11th generation" in line:
        if "core i3" in line:
            return "Mid Tier"
        elif "core i5" in line:
            return "Mid Tier"
        elif "core i7" in line:
            return "Mid Tier"
        elif "core i9" in line:
            return "High Tier"
    else:
        return "Other"
```

```python
def group_socket_amd(socket_type):
    '''

    socket_type: CPU socket type (str)
    return: type of socket based on usage (str)

    Given an AMD processor, its socket is grouped into types based on functionality.

    '''
    socket_type = socket_type.lower()
    if socket_type in ['am4']:
        return 'LGA'
    elif socket_type in ['sp3', 'fp4', 'fp5', 'fp6', 'g34', 'c32', 'ft3', 'ft3b', 'ft4', 'ft5']:
        return 'BGA'
    elif socket_type in ['am3+', 'am3', 'fm2+', 'fm2', 'am1', 'am2+', '939']:
        return 'FCPGA'
    elif socket_type in ['strx4', 'swrx8']:
        return 'Threadripper'    
    else:
        return 'Other'

def group_socket_intel(socket_type):
    '''

    socket_type: CPU socket type (str)
    return: type of socket based on usage (str)

    Given an Intel processor, its socket is grouped into types based on functionality.

    '''
    socket_type = socket_type.lower()
    if socket_type in ['fclga1151', 'fclga1150', 'fclga1155', 'fclga1156', 'fclga2011', 'fclga2011-3', 'fclga3647', 'fclga4189', 'fclga1200', 
                       'fclga2066']:
        return 'LGA'    
    elif socket_type in ['fcbga1283', 'fcbga1364', 'fcbga1168', 'fcbga1356', 'fcbga1440', 'fcbga1170', 'fcbga1296', 'fcbga1090', 'fcbga1515', 
                         'fcbga1493', 'fcbga1528', 'fcbga1338', 'fcbga1449', 'fcbga1344', 'fcbga1598', 'fcbga1787', 'fcbga1667', 'fcbga2518']:
        return 'BGA'    
    elif socket_type in ['fcpga946', 'fcpga988']:
        return 'FCPGA'    
    else:
        return 'Other'
```

```python
def group_memory(memory_type):
    '''

    memory_type: memory type (str)
    return: properly formatted memory type

    Given a string of a processor's memory type, the properly formatted string of the
    memory is returned.
    
    '''
    memory = memory_type.lower()
    if 'ddr3' in memory:
        return 'DDR3'
    elif 'lpddr4' in memory:
        return 'LPDDR4'
    elif 'lpddr3' in memory:
        return 'LPDDR3'
    elif 'ddr4' in memory:
        return 'DDR4'
    elif 'ddr2' in memory:
        return 'DDR2'
    else:
        return 'Other'
```

<span style="color:blue"><b>Now apply our functions to the AMD dataset to properly group our categorical columns.</b></span>

```python
amd = amd.dropna(subset=['product_line'])
amd["processor_performance"] = amd["product_line"].apply(group_product_line_amd)

amd = amd.dropna(subset=['socket'])
amd['socket_group'] = amd['socket'].apply(group_socket_amd)

amd = amd.dropna(subset=['memory_type'])
amd['memory_group'] = amd['memory_type'].apply(group_memory)
```

<span style="color:blue"><b>Next we will look at our Intel dataset.</span></b>

```python
intel
```

<span style="color:blue"><b>Now apply our functions to the Intel dataset to properly group our categorical columns.</b></span>

```python
intel = intel.dropna(subset=['socket'])
intel['socket_group'] = intel['socket'].apply(group_socket_intel)

intel = intel.dropna(subset=['memory_type'])
intel['memory_group'] = intel['memory_type'].apply(group_memory)

intel = intel.dropna(subset=['product_line'])
intel['processor_performance'] = intel['product_line'].apply(group_product_line_intel)
```

<span style="color:blue"><b>Since all the categorical features we want our now properly formatted, we will drop any unneccessary features based on our correlation table and domain knowledge. Additionally, features such as 'launch_date', are dropped because of very few entries.</b></span>

```python
amd = amd.drop(['id', 'threads', 'launch_date', 'turbo_frequency', 'socket', 'memory_type', 'url', 'vertical_segment',
               'product_line', 'name', 'max_memory_speed'], axis=1)
intel = intel.drop(['id', 'name', 'processor_number', 'threads', 'launch_date', 'turbo_frequency', 'socket', 'memory_type', 'url', 'vertical_segment',
                   'bus_speed', 'configurable_tdp_up_frequency', 'fullname', 'configurable_tdp_up', 'max_memory_size', 'status', 'package_size',
                   'price', 'product_line'], axis=1)
```

<span style="color:blue"><b>Next, we will fill our missing data in our numerical columns based off the skew of each respectively. To start, let's plot each columns frequency to see their distribution.</b></span>

```python
for i in intel.columns[:7]:
    plt.hist(intel[i], bins=20, color='skyblue', edgecolor='black')
    plt.title(i)
    plt.show()

for i in intel.columns[:7]:
    skew = intel[i].skew()
    print(f'{i} skewness: {skew}')
```

```python
for i in amd.columns:
    nan_counts = amd[i].isna().sum()
    print(f'{i} missing counts: {nan_counts}')
```

<span style="color:blue"><b>For imputting missing values, the rule of thumb used was a largely skewed column would use the median as a replacement or if slightly skewed, the mean would be used as a replacement.</b></span>

```python
intel['cores'] = intel['cores'].fillna(intel['cores'].median())
intel['lithography'] = intel['lithography'].fillna(intel['lithography'].median())
intel['tdp'] = intel['tdp'].fillna(intel['tdp'].median())
intel['max_temp'] = intel['max_temp'].fillna(intel['max_temp'].median())
```

<span style="color:blue"><b>Now let's do this for the AMD dataset.</b></span>

```python
for i in amd.columns[:7]:
    plt.hist(amd[i], bins=20, color='skyblue', edgecolor='black')
    plt.title(i)
    plt.show()

for i in amd.columns[:7]:
    skew = amd[i].skew()
    print(f'{i} skewness: {skew}')
```

```python
for i in amd.columns:
    nan_counts = amd[i].isna().sum()
    print(f'{i} missing counts: {nan_counts}')
```

```python
amd['cores'] = amd['cores'].fillna(amd['cores'].median())
amd['lithography'] = amd['lithography'].fillna(amd['lithography'].mean())
amd['tdp'] = amd['tdp'].fillna(amd['tdp'].mean())
amd['max_temp'] = amd['max_temp'].fillna(amd['max_temp'].mean())
```

<span style="color:blue"><b>Before we combine both datasets into one, we will manually 'encode' our datasets so that we have an additional feature for the brand of processor.</b></span>

```python
amd['brand'] = 0
intel['brand'] = 1

intel = intel[amd.columns]
amd.reset_index(drop=True, inplace=True)
intel.reset_index(drop=True, inplace=True)
total_df = pd.concat([amd, intel], ignore_index=True)
```

```python
for i in total_df.columns:
    nan_counts = total_df[i].isna().sum()
    print(f'{i} missing counts: {nan_counts}')

print()
skew = total_df['base_frequency'].skew()
print(skew)
```

```python
total_df['base_frequency'] = total_df['base_frequency'].fillna(total_df['base_frequency'].mean())
```
